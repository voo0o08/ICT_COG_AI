### 2. 인구수 대비 CCTV의 비율을 구하고, 그래프로 그려보자.
CCTV비율  이라는 칼럼을 data_result에 하나 추가.. 하고 그래프로 그리기.

```python
data_result['CCTV비율'] = data_result['총계'] / data_result['인구수'] * 100
```

```python
data_result.head()
```

```python
data_result['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
### Scatter 그래프로 인구수, CCTV 관계
plt.figure( figsize=(6,6))
plt.scatter( data_result['인구수'] , data_result['총계'], s=50 )
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
### 경향을 파악하기 위한 직선 하나를 추가해 봅시다.
fp1 = np.polyfit(data_result['인구수'], data_result['총계'], 1 )
```

```python
fp1
```

```python
f1 = np.poly1d( fp1 )
fx = np.linspace(100000, 700000, 100)
```

```python
fx
```

```python
plt.figure( figsize=(6,6))
plt.scatter( data_result['인구수'] , data_result['총계'], s=50 )
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

### 3. 오차값 구하기.
녹색 그래프와 실제 값의 차이를 오차 정의하자!  
data_result에 오차 칼럼을 추가해 보자!  

```python
data_result['오차']=np.abs( data_result['총계'] - f1(data_result['인구수']) )
```

```python
data_result.head()
```

```python
df_sort = data_result.sort_values('오차', ascending=False)
```

```python
plt.figure( figsize=(14,10))
plt.scatter( data_result['인구수'] , data_result['총계'], s=data_result['외국인비율']*50 , c=data_result['오차'])
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n] * 1.02, df_sort['총계'][n] * 0.98, df_sort.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.colorbar()
plt.grid()
plt.show()
```
